#### Description of the dataset

- **Job ID**: Unique identifier for each job posting (usually a numeric value)
  
- **Title**: Title of the job being offered
  
- **Location**: Geographical location where the job is based
  
- **Department**: The department within the company offering the job
  
- **Salary Range**: The range of salary offered for the job
  
- **Company Profile**: Brief description of the company
  
- **Description**: Detailed description of the job role
  
- **Requirements**: The set of skills, qualifications, and experience required
  
- **Benefits**: Additional perks or benefits offered with the job
  
- **Telecommuting**: Binary value indicating if telecommuting is allowed (1 for Yes, 0 for No)
  
- **Has Company Logo**: Binary value indicating if the job posting includes the company's logo (1 for Yes, 0 for No)
  
- **Has Questions**: Binary value indicating if there are screening questions (1 for Yes, 0 for No)
  
- **Employment Type**: Type of employment (Full-time, Part-time, Contract, etc.)
  
- **Required Experience**: Level of experience required for the job (Entry Level, Mid-Senior, etc.)
  
- **Required Education**: Minimum educational qualification required
  
- **Industry**: The industry category the job falls under
  
- **Function**: The specific function or role of the job within the company
  
- **Fraudulent**: Binary value indicating if the job posting is fraudulent (1 for Yes, 0 for No)

#### Dealing with variables

```python
for variable in df.columns:
    print(f"Unique Values in {variable} :", len(df[variable].unique()))
```

```python
def compare_with_fraudulent(column_name, data=df):
    # Increase the figure size
    plt.figure(figsize=(15,8))

    # Order bars based on frequency
    order = data[column_name].value_counts().index
    
    # Create the countplot
    sb.countplot(data=data, x=column_name, hue='fraudulent', order=order)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)

    # Set title and display the plot
    title = f"{column_name} vs Fraudulent"
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
compare_with_fraudulent('telecommuting')
```

```python
# Plot the countplot
compare_with_fraudulent('has_questions')
```

```python
# Plot the countplot
compare_with_fraudulent('employment_type')
```

```python
# Plot the countplot
compare_with_fraudulent('required_experience')
```

```python
compare_with_fraudulent('required_education')
```

```python
df.fillna(' ', inplace=True)
def split(location):
    l = location.split(',')
    return l[0]

df['country'] = df.location.apply(split)
```

```python
# Get the number of unique values for each column
unique_counts = df.nunique().sort_values(ascending=False)

# Display the columns with the most unique values
print(unique_counts)
```

```python
df['text'] = df['description'] + " " + \
                      df['requirements'] + " " + \
                      df['title'] + " " + \
                      df['benefits'] + " " + \
                      df['location'] + " " + \
                      df['company_profile']
```

```python
non_fraud_text = ' '.join(df[df['fraudulent'] == 0]['text'].astype(str))
fraud_text = ' '.join(df[df['fraudulent'] == 1]['text'].astype(str))
```

```python
wordcloud1 = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = set(STOPWORDS), 
                min_font_size = 10).generate(non_fraud_text)

plt.figure(figsize = (8, 8)) 
plt.imshow(wordcloud1) 
plt.title("Non-Fraudulent Postings")
plt.axis("off")
plt.show()
```

```python
wordcloud2 = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = set(STOPWORDS), 
                min_font_size = 10).generate(fraud_text)
plt.figure(figsize = (8, 8)) 
plt.imshow(wordcloud2) 
plt.title("Fraudulent Postings")
plt.axis("off")
plt.show()
```
